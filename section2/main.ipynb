## 確率分布を示す数式

```python
# 正規分布関数の定義
def norm(x, mu, sigma):
    return np.exp(-((x - mu)/sigma) **2/2) / (np.sqrt(2*np.pi) * sigma)

# パラメータ定義
mu1, sigma1 = 3.0, 2.0
mu2, sigma2 = 1.0, 3.0
# グラフ描画用x座標の定義
# 2つの正規分布関数で±3σの範囲を描画
x = np.arange(-8, 10, 0.01)

# x軸目盛りの設定
xticks = np.arange(-8, 11, 1)

# グラフ描画
plt.plot(x, norm(x, mu1, sigma1), label=f'mu={mu1}, sigma={sigma1}')
plt.plot(x, norm(x, mu2, sigma2), label=f'mu={mu2}, sigma={sigma2}')
plt.xticks(xticks, fontsize=12)
plt.legend(fontsize=12)
plt.title("正規分布の確率密度関数")
```

## 一様分布
- 区間[0,1]の一様乱数を生成。
- この分布は3章以降で取り上げる確率モデル定義時の事前分布として用いられることが多い。
- 確率値を確率変数とする確率分布を考える場合に、
  - どういう種類の確率分布であるのか、さらにその確率分布のパラメータに関する手がかりが事前に一切ないとする。
  - この場合、「０から１までの値の可能性は全て同等に確からしい」とするのが一つの考え方。
     - この考え方を確率分布に当てはめると一様分布になる。
     - ※　自然界で確率変数が一様分布になる現象はないが。

```python
# 一様分布の確率モデル定義
lower = 0.0
upper = 1.0

model6 = pm.Model()
with model6:
    # pm.Uniformで一様分布を指定
    # lowerは下限値
    # upperは上限値
    x = pm.Uniform("x", lower=lower, upper=upper)


with model6:
    # サンプリング
    prior_samples6 = pm.sample_prior_predictive(random_seed=42)
  
# サンプル値の抽出
x_samples6 = prior_samples6["prior"]['x'].values
print(x_samples6[:,:100])

# サンプリング結果の可視化
ax = az.plot_dist(x_samples6)
ax.set_title(f"一様分布の事前分布のトレースプロット lower = {lower}, upper = {upper}")
```

```python
# サンプリング結果のヒストグラム表示
bins = np.arange(0, 1.1, 0.1)
ax = az.plot_dist(x_samples6, kind='hist', hist_kwargs={'bins':bins})
ax.set_title(f"一様分布の事前分布のヒストグラム lower = {lower}, upper = {upper}")

summary6 = az.summary(prior_samples6,kind='stats')
display(summary6)
```

```python
# 一様分布の確率モデル定義

# パラメータ設定
lower = 0.1
upper = 0.9

model7 = pm.Model()

with model7:
  # pm.Uniformで一様分布を指定
  # lowerは下限値
  # upperは上限値
  x = pm.Uniform("x", lower=lower, upper=upper)
  
  # サンプリング
  prior_samples7 = pm.sample_prior_predictive(random_seed=42)

# サンプル値抽出
x_samples7 = prior_samples7["prior"]['x'].values
print(x_samples7[:,:100])

# サンプリング結果の可視化
ax = az.plot_dist(x_samples7)
ax.set_title(f"一様分布の事前分布のトレースプロット lower = {lower}, upper = {upper}")
```

## ベータ分布
- 正規分布と同様の連続分布
- 正規分布が実世界に対応がつきやすいのに対して、ベータ分布は統計学の知識が前提になるためわかりにくい。
- ※ 4章の実習後改めて読み直すと良いよ

### 対応する事象
- ベータ分布には、正規分布でいう「試験の得点分布」の様な実世界で直接対応づくような事例がない。（ベータ分布がわかりにくい理由）
 - 二項分布を考えた時
  $$
    P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
  $$

  - ここで、
    - \( P(X = k) \) は、試行回数 \( n \) のうち \( k \) 回成功する確率。
    - \( \binom{n}{k} \) は \( n \) 回の試行の中で \( k \) 回成功する組み合わせの数（コンビネーション）。
    - \( p \) は1回の試行で成功する確率。
    - \( 1 - p \) は1回の試行で失敗する確率。
 - この時、n = 5, k =2 のケースで考えると、上の式は
  $$
   P(X = 2) = 10 p^2 (1 - p)^{3}
  $$

 - 元々は、kの関数だったが、特定のkの値に固定することで「pの関数」と見ることができる。

 - この「pの関数」の意味は、「くじ引きを5回引いて2回当たりだった。この時、くじに当たる確率がpであることはどの程度確からしいか」というpを確率変数とする連続分布になる。
  - = これがベータ分布
  - 確率値を目的変数とした確率分布


## 確率分布を示す数式
- 上の具体例を引き継いで、確率分布を示す式を作ってみる
$$
   f(p) = C p^{2} (1 - p)^{3}
$$

- 上式のCは、上の数式が確率密度関数になるための調整すべき定数で、上の関数の0から1までの面積がちょうど1になるように決める。

- 具体的なCの値は以下に示す。
   - 今、くじに当たった回数をs、外れた回数をfとすると、
   - α=s+1, β=f+1で2つのパラメータを定める。（この分布関数の平均値は α/(α+β)と知られており、数学的に1を加えたほうが綺麗に説明できるらしい）

$$
   f(p) = C p^{α-1} (1 - p)^{β-1}
$$

$$
   C = \frac{(\alpha + \beta - 1)!}{(\alpha - 1)! (\beta - 1)!}
$$

```python
from math import gamma
#パラメータ定義
alpha = 2.0 + 1.0
beta = 3.0 + 1.0

# ベータ関数の定義
def Beta(p, alpha, beta):
  C = gamma(alpha + beta - 1) / (gamma(alpha -1) * gamma(beta -1))
  return C * p ** (alpha -1) * (1-p) ** (beta -1)

# グラフ描画用x座標の定義
# 0.0 < p < 1.0の範囲で0.01刻み
p = np.arange(0.0, 1.0, 0.01)

# グラフ描画
plt.plot(p, Beta(p, alpha, beta))
plt.title(f"ベータ分布の確率密度関数 alpha = {alpha}, beta = {beta}")
plt.xlabel(r'$p$')
plt.ylabel('尤度')
```

- このグラフの縦軸は、くじに当たる確率pの確からしさ（尤度）を表していて、
 - p = 0.4が最も確からしい確率値であることをグラフは示している
 - くじに当たった回数2回、外れた回数3回なので当たる確率0.4では？の直感的予想と合っている

```python
# 確率モデル定義

# パラメータ設定
alpha = 2.0 + 1.0
beta = 3.0 + 1.0

model8 = pm.Model()
with model8:
    # pm.Betaでベータ分布を指定
    # alphaは事前分布のパラメータ くじに当たった回数+1
    # betaは事前分布のパラメータ くじに外れた回数+1
    x = pm.Beta("x", alpha=alpha, beta=beta)


# サンプリング
with model8:
    prior_samples8 = pm.sample_prior_predictive(random_seed=42)
  

# サンプル値の抽出
p_samples8 = prior_samples8["prior"]['x'].values
# 桁数が多いので先頭100個のみに限定
print(p_samples8[:,:100])

# サンプリング結果の可視化
ax = az.plot_dist(p_samples8)
ax.set_title(f"ベータ分布の事前分布のトレースプロット alpha = {alpha}, beta = {beta}")
```

```python
# ヒストグラム表示と統計情報表示
bins = np.arange(0, 1.0, 0.1)

ax = az.plot_dist(p_samples8, kind='hist', hist_kwargs={'bins':bins})
ax.set_title(f"ベータ分布の事前分布のヒストグラム alpha = {alpha}, beta = {beta}")

summary8 = az.summary(prior_samples8,kind='stats')
display(summary8)
```

### ベータ分布と一様分布の関係
$$
   f(p) = C p^{α-1} (1 - p)^{β-1}
$$

というベータ分布の式で、α=1, β=1の特殊なケースを考えてみる。

- すると、f(p)が定数になる。
- =「1回も試行していない特殊な二項分布に対応した確率値の確率分布は一様分布である」と解釈できる。
- =「ベータ分布の特殊なケースが一様分布である」とも言える

## 半正規分布
- 対応する事象
 - 正規分布で標準偏差の値を確率変数で表現することを考える。
 - 標準偏差の値は負の値を取らない。
   - 標準偏差は分散の平方根だから
 - そこで事前分布に対しても、「正の値のみをとる」という条件を満たす確率分布の確率変数が望ましいことになる。
   - 確率変数として考える標準偏差が従う分布の事前分布

== 半正規分布は、このような制約を満たす確率変数を定義する際によく用いられる
- ベイズ推論で確率モデルを構築するにあたり事前分布として便宜上用いられる確率分布（そのため、実世界との対応づけは意識しなくて良い）
- ※本書では、分かりやすさ重視で正規分布の標準偏差の事前分布は全て半正規分布を用いている。
  - 逆ガンマ、ガンマ、レイリー分布など他にもあるらしい


### 確率分布を示す数式
- 半分に切り取る前の正規分布は、y軸に関する対称性を保つために平均μ=0の制約がつく。
- そのため、確率分布の数式におけるパラメータは標準偏差のσの値のみとなる。
- これを考慮した確率分布を示す数式は以下の通り
$$
f(x \mid \sigma) = 
\begin{cases} 
\frac{\sqrt{2}}{\sigma \sqrt{\pi}} \exp\left( -\frac{x^2}{2\sigma^2} \right) & \text{for } x \geq 0 \\
0 & \text{for } x < 0 
\end{cases}
$$


```python
# パラメータ設定
sigma = 1.0

model10 = pm.Model()
with model10:
  # pm.HalfNormalで半正規分布を指定
  # sigmaは標準偏差
  x = pm.HalfNormal("x", sigma=sigma)
  
  # サンプリング
  prior_samples10 = pm.sample_prior_predictive(random_seed=42)

# サンプル値の抽出
x_samples10 = prior_samples10["prior"]['x'].values
print(x_samples10[:,:100])

# サンプリング結果の可視化
ax = az.plot_dist(x_samples10)
ax.set_title(f"半正規分布の事前分布のトレースプロット sigma = {sigma}")
```

## コラム（HDI highest density interval）とCI（confidence interval）の違い
- これらは、端的にいうと「得られたサンプルデータを値の大きさ順に並べた時に、中心部分と周辺の部分を区別した際の中心部分の領域のこと」

- 2つの違いは「何を持って中心部分とするか」
  - CI:信頼区間
    - 統計学的にはこちらがポピュラー
    ー 「80％信頼区間」＝両端に対して、両端からそれぞれの面積が全体の10%になるようなxを求めることと同義

  - HDI
    - step1: yの値を定める
      - 確率密度関数のグラフとy=kの好転を求め、できた交点のx座標がx1,x2だとすると、x=x1,x=x2の直線を引く
      - そして、2つの直線に挟まれた部分の面積が全体の何%かを計算する。（kを変化させると変化）
      - ちょうど比率が80%になる場所(y=k)を見つける
    
    - step2:もとまったy=kのkの値に対応するxの範囲を求める。
      - この時できたxの範囲がHDI
      - ※ CIの場合、区間はグラフのx軸に関する両端から狭めていくので求められる区間は必ず１つに繋がっていく。
      - ※これに対して、HDIの場合、確率密度関数のピークが2箇所ある時、求められる区間が２つ以上に分かれる場合がある。
      - よって、理想的な正規分布の時以外は、CI・ HDIの結果は異なる。
    
    - まとめ
     - 「全体に対する面積比率がr%となるx座標の範囲」という点は同じだがアルゴリズムが異なり結果も異なる場合がある。
     - Arvizは標準的にHDIを採用しているため本書では「HDIによるx座標の範囲」の考え方を用いることとしている。
